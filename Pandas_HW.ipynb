```python
bins = [0,9,14,19,24,29,34,39,200]
```

```python
df["AgeBucket"] = pd.cut(df["Age"], bins)
df
```

```python
x = df.groupby('AgeBucket', as_index=False).SN.count()
x
```

```python
x["Pct_of_Players"] = x['SN']/df["SN"].count()
x
```

```python
age_demo = pd.DataFrame({'-': ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'],'Total Count': ['28','35','133','336','125','64','42','17'],'Percentage of Players': ['3.59','4.49','17.05','43.08','16.03','8.21','5.38','2.18']})
age_demo
```

```python
avg_purchase = df.groupby('AgeBucket', as_index=False).Price.mean()
avg_purchase
```

```python
total_purch = df.groupby('AgeBucket', as_index=False).Price.sum()
total_purch
```

```python
avgUser = df.groupby('SN', as_index=False).Price.mean()
avgUser

```

```python
###avg of avg prices
want = df[['AgeBucket','SN']].drop_duplicates()
want 
avgus1 = pd.merge(avgUser, want, how='left', on='SN').drop_duplicates()

```

```python
###avg of avg by bucket
avgUseravg = avgus1.groupby('AgeBucket', as_index=False).Price.mean()
avgUseravg
```

```python
###final age bucket analysis 
ll=pd.merge(x,avg_purchase, how='left', on=['AgeBucket'])
lll=pd.merge(ll,total_purch,how='left' , on='AgeBucket')
llll=pd.merge(lll,avgUseravg,how='left',on='AgeBucket')
llll
```

```python
###final bin  analysis
bin_final =  pd.DataFrame({"-": llll['AgeBucket'],"Purchase Count": llll['SN'],"Percent of Players":llll['Pct_of_Players'],"Average Purchase Price":llll['Price_x'],"Total Purchase Value":llll['Price_y'],"Avg Total Purchase per Person":llll['Price']})
bin_final
```

```python
#top buyers
topb = df.groupby('SN', as_index=False).Price.count()
topb
```

```python
#top spenders
tops = df.groupby('SN', as_index=False).Price.sum().sort_values('Price', ascending=False).head()
tops
```

```python
#average purch
avg_purch=df.groupby('SN', as_index=False).Price.mean().sort_values('Price', ascending=False)
avg_purch
```

```python
left=pd.merge(tops, topb, how='left', on=['SN'])
left
```

```python
leftfinal=pd.merge(left, avg_purch, how='left', on=['SN'])
leftfinal
```

```python
new_final =  pd.DataFrame({"SN": leftfinal['SN'],"Total Purchase Value": leftfinal['Price_x'],"Purchase Count":leftfinal['Price_y'],"Average Purchase Price":leftfinal['Price']})
new_final
```
